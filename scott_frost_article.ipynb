```python
nebraska['PredictedWins'] = nebraska.win_prob.rolling(12).sum()
nebraska['ActualWins'] = nebraska.W_L.rolling(12).sum()
nebraska.PredictedWins.plot(color='red', figsize = (8,6))
nebraska.ActualWins.plot(color='black', style = '--')

#nebraska.win_prob.plot()
plt.suptitle('A Story of Poor Play and Underperformance', fontweight ="bold")
plt.xlabel('Scott Frost Game Number at Nebraska')
plt.ylabel('Wins')
plt.legend()
#plt.savefig("my_plot.png")
plt.show()
```
